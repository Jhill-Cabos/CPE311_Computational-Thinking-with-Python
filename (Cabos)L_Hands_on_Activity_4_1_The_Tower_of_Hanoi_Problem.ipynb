In order to get transfer the disks, I must first follow the following rules:


*   Only one disk can be transfer at a time.
*   Each move consists of taking the upper disk from one of the peg and placing it on the top of another peg i.e. a disk can only be moved if it is the uppermost disk of the peg.



*   Never a larger disk is placed on a smaller disk during the transfer.




```python
def TOH(move, Sour, Aux, Des): #no. of disks, source, Auxiliary,  Destination
  if move == 1:
    print("Move Disk ", move, " from ", Sour, " to ", Des) #this is where the disk moves
  else:
    TOH(move-1,Sour, Des, Aux)#this is where the disk leaves
    print("Move Disk ",  move, " from ", Sour, " to ", Des)
    TOH(move-1,Aux,Sour,Des)

move = int(input("Enter no. of disks: ")) #number of disks
TOH(move,'A','B','C')
```

In conclusion, In this problem, we have three pegs (A, B, and C) and a set of n (or in mine it is 'move') disks of different sizes.
