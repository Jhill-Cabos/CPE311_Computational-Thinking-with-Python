**Coefficient of variation interquartile**



```python
cmean = mean(data)
cstdev = stdev(data)
CV = (cstdev / cmean) * 100

CV
```

```python
# Interquartile Range
q1 = quantiles(salaries, n=4)[0]
q3 = quantiles(salaries, n=4)[2]
IR = q3 - q1

(IR)
```

**Quartile coefficient of dispersion**

```python
qmean = (q1 + q3) / 2
qcd = (IR / qmean) * 100
(qcd)
```
